```python
import pandas as pd

df_neg = pd.read_csv('../final-dataset/negative.csv', usecols=['github', 'message'])

df_no_null = df_neg[pd.notnull(df_neg['message'])]
print(f"Found {len(df_no_null)} viable commits in {len(df_neg)} initial commits")

df_non_neg = df_no_null[~df_no_null['message'].str.contains('vuln|secur|attack|patch|sec.*bug|exploit|xss|mem.*leak|csrf|scripting', regex = True)]
print(f"Cleaning positive commits {len(df_no_null)} => {len(df_non_neg)}")

df_non_neg.to_csv('../final-dataset/negative_clean.csv', index=False)
```

```python
import pandas as pd

df_neg = (pd.read_csv('../final-dataset/negative_clean.csv', usecols=['github', 'message']))
df_neg = df_neg[pd.notnull(df_neg['message'])]
df_neg['type'] = 'negative'

for idx, row in df_neg.iterrows():
    df_neg.at[idx, 'sha'] = row['github'].split('/')[-1]
```

```python
df_pos = (pd.read_csv('../final-dataset/positive.csv', usecols=['github', 'message', 'sha']))
df_pos = df_pos[pd.notnull(df_pos['message'])]
df_pos['type'] = 'positive'
```

```python
df = pd.concat([df_pos, df_neg], axis=0)
```

```python
df.to_csv('../final-dataset/messages.csv', index=False)
```
